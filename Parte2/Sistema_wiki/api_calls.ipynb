```python
from pymongo import MongoClient
import json

client = MongoClient('mongodb://localhost:27017/')
db = client['dungeons_original']
```

GET /loot: Devuelve los siguientes campos de todos los objetos del juego: idL, name

```python
def get_loot():
    cursor = db.loot.find({}, {'_id': 0, 'id': 1, 'name': 1}).sort('id', 1)
    return json.dumps(list(cursor), indent=4)

print(get_loot())
```

GET /loot/{loot_id}: Devuelve toda la información de un objeto del juego, incluidas las
habitaciones donde aparece, de las habitaciones devuelve: idR, room_name y de la mazmorra a
la que pertenece: idM, dungeon_name.

```python
def get_loot_by_id(id):
    cursor = db.loot.find({'id': id}, 
                          {'_id': 0,
                            'id': 1,
                            'name': 1, 
                            'gold': 1, 
                            'type1': 1, 
                            'type2': 1, 
                            'weight': 1, 
                            'in_rooms.room_id': 1, 
                            'in_rooms.room_name': 1, 
                            'in_rooms.dungeon_id': 1, 
                            'in_rooms.dungeon_name': 1})
    return json.dumps(list(cursor), indent=4)

print(get_loot_by_id(53))
```

GET /dungeon: Devuelve los siguientes campos de todas las mazmorras del juego: idD, name.

```python
def get_dungeon():
    cursor = db.rooms.aggregate( [
        {"$group": { "_id": {'dungeon_id': "$dungeon_id", 'dungeon_name': "$dungeon_name"}}},
        {"$sort": {"_id.dungeon_id": 1}}, 
        {"$project": {"_id": 0, 
                      "idD": "$_id.dungeon_id", 
                      "name": "$_id.dungeon_name"}}
                      ])
    
    return json.dumps(list(cursor), indent=4)

print(get_dungeon())
```

GET /dungeon/{dungeon_id}: Devuelve información sobre una mazmorra del juego. Debe
devolver: idM, name y lore. Además, este endpoint se utiliza para alimentar un grafo interactivo
por lo que requiere la siguiente información: 1) el nombre e id de cada habitación de la
mazmorra; 2) las conexiones entre habitaciones de la mazmorra; 3) el id y el nombre de los
monstruos que aparecen en cada habitación; 4) el id y el nombre de los tesoros que aparecen en
cada habitación; 5) El número de comentarios de cada categoría que hay en cada habitación.

```python
def get_dungeon_by_id(id):
    response = {}
    dungeon_info = db.rooms.find_one({'dungeon_id': id}, {'_id': 0, 'dungeon_id': 1, 'dungeon_name': 1})
    response['dungeon_id'] = dungeon_info['dungeon_id']
    response['dungeon_name'] = dungeon_info['dungeon_name']
    response['rooms'] = []

    rooms = db.rooms.find({'dungeon_id': id}, {'_id': 0,
                                                'room_id': 1,
                                                'room_name': 1, 
                                                'rooms_connected': 1,
                                                'in_waypoint': 1, 
                                                'out_waypoint': 1}
                                                
    ).sort('room_id', 1)
    for room in list(rooms):
        id = room['room_id']
        count_of_comment_categories = db.rooms.aggregate([
            {"$match": {"room_id": id}},
            {"$unwind": "$hints"},
            {"$group": {"_id": "$hints.category", "count": {"$sum": 1}}},
            {"$project": {"_id": 0, "category": "$_id", "count": 1}}
        ])
        count_of_monsters = db.rooms.aggregate([
            {"$match": {"room_id": id}},
            {"$unwind": "$monsters"},
            {"$group": {"_id": {'id':"$monsters.id", 'name':'$monsters.name'}, "count": {"$sum": 1}}},
            {"$project": {"_id": 0, "monster_id": "$_id.id", "monster_name": "$_id.name", "count": 1}}
        ])
        
        count_of_loot = db.rooms.aggregate([
            {"$match": {"room_id": id}},
            {"$unwind": "$loot"},
            {"$group": {"_id": {'id':"$loot.id", 'name':'$loot.name'}, "count": {"$sum": 1}}},
            {"$project": {"_id": 0, "loot_id": "$_id", "count": 1}}
        ])
        
        room['hints_distribution'] = list(count_of_comment_categories)
        room['monsters'] = list(count_of_monsters)
        room['loot'] = list(count_of_loot)
        response['rooms'].append(room)
    
    return json.dumps(response, indent=4)

print(get_dungeon_by_id(0))
```

GET /room/{room_id}: Devuelve la siguiente información de una habitación: idR, name, inWP y
outWP. Además, incluye la siguiente información de todos los monstruos incluidos en la
habitación (idM, name, type, level, place, exp, manPage) y de los tesoros (idL, name, type1,
type2, weight, gold). Por último, incluye todos los comentarios que se hayan realizado para esa
habitación, cada comentario debe incluir: email y userName, country, creationDate del usuario
que lo realizo y texto, fecha de publicación y categoría del comentario.

```python
def get_room_by_id(id):
    cursor = db.rooms.find({'room_id': id}, 
                          {'_id': 0,
                            'room_id': 1,
                            'room_name': 1, 
                            'in_waypoint': 1, 
                            'out_waypoint': 1, 
                            'monsters.id': 1,
                            'monsters.name': 1,
                            'monsters.type': 1,
                            'monsters.level': 1,
                            'monsters.place': 1, 
                            'monsters.man_page': 1,
                            'loot.id': 1,
                            'loot.name': 1,
                            'loot.gold': 1,
                            'loot.type1': 1,
                            'loot.type2': 1,
                            'loot.weight': 1,
                            'hints.category': 1,
                            'hints.hintText': 1,
                            'hints.creation_date': 1,
                            'hints.publish_by.email': 1,
                            'hints.publish_by.user_name': 1,
                            'hints.publish_by.country': 1,
                            'hints.publish_by.creation_date': 1})

    
    return json.dumps(list(cursor), indent=4)

print(get_room_by_id(5))
```

GET /user: Devuelve los campos email, username y country de todos los usuarios.

```python
def get_user():
    cursor = db.users.find({}, {'_id': 0, 'email': 1, 'user_name': 1, 'country': 1}).sort('email', 1)
    return json.dumps(list(cursor), indent=4)

print(get_user())
```

GET /user/{email}: Devuelve todos los campos de un usuario. Además, incluye todos los
comentarios que ha realizado. De cada comentario incluye, el texto, la fecha de creación, la
categoría, el id (Room.IdR) y nombre (Room.name) de la habitación a la que hace referencia el
comentario, el id(Dungeon.IdD) y nombre(Dungeon.name) de la mazmorra donde está la
habitación.

```python
def get_user_by_email(email):
    cursor = db.users.find({'email': email}, 
                          {'_id': 0,
                            'email': 1,
                            'user_name': 1, 
                            'country': 1, 
                            'creation_date': 1,
                            'hints.category': 1,
                            'hints.text': 1,
                            'hints.creation_date': 1,
                            'hints.referemces_room.room_id': 1,
                            'hints.referemces_room.room_name': 1,
                            'hints.referemces_room.dungeon_id': 1,
                            'hints.referemces_room.dungeon_name': 1
                            })

    return json.dumps(list(cursor), indent=4)

print(get_user_by_email('aabe@example.net'))
```
