```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['dungeons']
```

```python
import json
def get_loot():
    cursor = db.loot.find({}, {'_id': 0, 'id': 1, 'name': 1}).sort('id', 1)
    return json.dumps(list(cursor), indent=4)
print(get_loot())
```
