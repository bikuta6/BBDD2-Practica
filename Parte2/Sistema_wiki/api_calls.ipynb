```python
from pymongo import MongoClient
import json

client = MongoClient('mongodb://localhost:27017/')
db = client['dungeons']
```

```python
def get_loot():
    cursor = db.loot.find({}, {'_id': 0, 'id': 1, 'name': 1}).sort('id', 1)
    return json.dumps(list(cursor), indent=4)

def get_loot_by_id(id):
    cursor = db.loot.find({'id': id}, {'_id': 0}).sort('id', 1)
    return json.dumps(list(cursor), indent=4)
```

```python
def get_monsters():
    cursor = db.monster.find({}, {'_id': 0, 'id': 1, 'name': 1, 'level': 1, 'type': 1}).sort('id', 1)
    return json.dumps(list(cursor), indent=4)

def get_monster_by_id(id):
    cursor = db.monster.find({'id': id}, {'_id': 0}).sort('id', 1)
    return json.dumps(list(cursor), indent=4)

print(get_monsters())
```
